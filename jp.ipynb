$$\ddot{x} + \alpha (x^2 - 1 ) \dot{x} + x = 0$$
lo que importa aca es como esta tomada la ecuacion!
$$\frac{\partial F}{\partial x} = 0 $$
$$F_{xx} + F_{tt} = 0$$

Ahora lo que voy hacer es agregar los metodos numericos (Euler y RK4)

Euler:

Sea $$\frac{dy}{dx}=f(y(x),x), y(x_0)=y_0$$
Si se realiza la aproximacion $f(y(x),x) \thickapprox \frac{y(x+h)-y(x)}{h}$ se obtiene la siguiente relacion para valores discretizados 
$$y_{n+1} \thickapprox \Delta_X f(y_n,x_n) + y_n$$
Con $x_n = x_0 + \Delta_x * n$ y $N=\frac{x_{N-1}-x_0}{\Delta_x}+1$


```python
#sea el ejemplo para el caso de f(y(x), x) = y*x
delta = 0.001
x0=0
xF=1
x = np.arange(x0, xF+delta, delta)
#N= ((xF-x0)/delta)+1
N=len(x)
y = np.zeros((N))
y[0]= 2

for i in range(0, (N-1)):
    y[i+1] = delta * (y[i]*x[i]) + y[i]

#mientras que la solucion real
yR = 2* np.exp((x**2)/(2))

plt.plot(x,y,'.',color="red")
plt.plot(x, yR, color="darkviolet")
plt.grid(True)
```
